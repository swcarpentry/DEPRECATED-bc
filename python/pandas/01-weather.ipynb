```python
import pandas
pandas.set_option('display.mpl_style', 'default')
```

```python
df = pandas.read_csv('./weather_2012_mtl.csv', index_col='Date/Time', parse_dates=True)
```

We can look at the first 5 rows of our dataframe like this:

```python
df[:5]
```

We select a column from our dataframe like this:

```python
df['Temp (C)']
```

```python
%matplotlib inline
```

```python
df['Temp (C)'].plot(figsize=(20, 5))
```

What if we wanted the average temperature for each month? This is one line of pandas code:

```python
df['Temp (C)'].resample('M', how='mean').plot(kind='bar', figsize=(20, 5))
```

It turns out that `df[300:400]` is all dates from Jan. 13 at noon to Jan. 17 at 3pm. Guessing and checking is time-consuming, though. Pandas lets us take slices using dates:

```python
df['2012-06':'2012-09']['Temp (C)']
```

```python
df['2012-06':'2012-09']['Temp (C)'].plot(figsize=(20, 5))
```

What if we wanted to plot the minimum, median, and maximum temperatures on the same graph?

```python
minimums = df['Temp (C)'].resample('M', how='min')
maximums = df['Temp (C)'].resample('M', how='max')
medians = df['Temp (C)'].resample('M', how='median')
```

```python
monthly_temps = pandas.DataFrame({'min': minimums, 'max': maximums, 'median': medians})
```

```python
monthly_temps.plot()
```
