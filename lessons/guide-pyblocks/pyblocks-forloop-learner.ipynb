# Basic Programming Using Python: Loops (Learner Version)

## Objectives

-   Repeat operations using a `for` loop.
-   Trace the value of a loop variable.
-   Use the `range` function to loop over a range of integers.

## Lesson

* Print vowels the hard way

```python
print 'a'
print 'e'
print 'i'
print 'o'
print 'u'
```

* Now print them the easy way using a loop

```python
for vowel in "aeiou":
    print vowel
```

* Update a variable repeatedly

```python
result = ""
for vowel in "aeiou":
    result = result + 2 * vowel
print result
```

* Use indentation to show what's in the body of the loop

```python
result = ""
for vowel in "aeiou":
    result = result + 2 * vowel
    print result
```

* Use the `range` function to generate the numbers 0, 1, 2, ...

```python
for i in range(3):
    print i
```

* Use a loop to operate on many cells of a grid

```python
from ipythonblocks import ImageGrid
grid = ImageGrid(10, 5)
for i in range(10):
    grid[i, 0] = (255, 0, 0)
grid.show()
```

* Make sure the indices are in the right order

```python
for i in range(10):
    grid[0, i] = (255, 0, 0)
```

```python
grid.show()
```

* Better to use grid's properties as bounds on loop
* DRY: Don't Repeat Yourself

```python
for i in range(grid.height):
    grid[0, i] = (255, 0, 0)
```

* Color a sub-rectangle the hard way

```python
for i in range(3):
    grid[0, i] = (0, 0, 255)
for i in range(3):
    grid[1, i] = (0, 0, 255)
for i in range(3):
    grid[2, i] = (0, 0, 255)
grid.show()
```

* Easier to use a nested loop

```python
for x in range(3):
    for y in range(3):
        grid[x, y] = (0, 0, 255)
grid.show()
```

* Inner loop's bounds can depend on current value of outer loop's variable

```python
grid = ImageGrid(4, 4)
for x in range(4):
    for y in range(x):
        grid[x, y] = (0, 255, 0)
grid.show()
```

```python
for y in range(4):
    for x in range(y):
	grid[x, y] = (0, 0, 255)
grid.show()
```

```python
for i in range(4):
    grid[i, i] = (0, 255, 255)
grid.show()
```

* Using color names is better than using raw RGB values

```python
from ipythonblocks import ImageGrid, colors
grid = ImageGrid(5, 5)
for x in range(grid.width):
    for y in range(x):
        grid[x, y] = colors['Lime']
for y in range(grid.height):
    for x in range(y):
        grid[x, y] = colors['Blue']
for i in range(grid.height):
    grid[i, i] = colors['Aqua']
grid.show()
```
