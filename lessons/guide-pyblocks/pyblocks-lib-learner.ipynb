## Lesson

* Use `import` to load a library, and `library.function` to refer to a function from it

```python
import math
print "pi is", math.pi
print "square root of 5 is", math.sqrt(5)
```

* Several ways to import things from libraries

```python
from math import sqrt
print sqrt(3)
```

```python
from math import hypot as euclid
print euclid(3, 4)
```

```python
from math import *
print sin(pi)
```

* Use `ipythonblocks` library to create grids of colored squares
* Use `object.method` to call a method (a function that belongs to a particular object)

```python
from ipythonblocks import ImageGrid
grid = ImageGrid(10, 5)
grid.show()
```

* Objects also have data members

```python
print 'width:', grid.width
print 'height:', grid.height
print 'lines_on:', grid.lines_on
```

* Some members are mutable, others are not

```python
grid.lines_on = False
grid.show()
```

```python
grid.width = 100
```

* Use `help(thing)` to get help

```python
help(grid)
```

* Use RGB (red-green-blue) triples to represent colors
* Each value must be in 0..255 (the range that can be represented in a single byte)
* Use `show_color` to display colors

```python
from ipythonblocks import show_color
show_color(255, 0, 0) # all reduce
```

```python
show_color(0, 255, 0)
show_color(0, 0, 255)
```

```python
show_color(50, 100, 200)
```

* Refer to a particular cell of a grid using an (x, y) index
* (0, 0) is the lower-left corner

```python
grid[0, 0] = (0, 255, 0)
grid.show()
```

```python
from ipythonblocks import BlockGrid
blocks = BlockGrid(6, 3)
blocks[1, 2] = (255, 0, 0)
blocks.show()
```

## Key Points

- A library (or module) is a collection of functions and variables that can be loaded into other programs.
- Most of the power of a programming system is in its libraries, rather than in the language itself.
- We refer to the things in a library using dotted notation.
- We can represent colors using RGB triples, in which each color component value is between 0 and 255.
- We can use `ipythonblocks.ImageGrid` to draw rectangular grids of cells.
- Grids have methods (functions) and members (variables), which we refer to using dotted notation.
- We refer to a specific cell in a grid using `g[x, y]`.
